#Linear regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.linear_model as lm
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/sahilaf/Data/refs/heads/main/canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.scatter(x=df['year'], y=df['per capita income (US$)'])
plt.xlabel('Year')
plt.ylabel('Income')
plt.show()
```

```python
reg = lm.LinearRegression()
reg.fit(df[['year']], df['per capita income (US$)'])
```

```python
reg.predict(np.array([[2016]]))
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
828.46507522*2016+(-1632210.7578554575)
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Income')
plt.scatter(df['year'], df['per capita income (US$)'])

# Generate points for the line
x = np.array(df['year'])  # Use existing years in your DataFrame
y_pred = reg.predict(x.reshape(-1, 1))  # Predict income for those years

# Plot the regression line
plt.plot(x, y_pred, color='red', linewidth=2)  # Customize color, width, etc.

plt.show()
```
