
# Dataset preparation (random split into training, validation, and test sets)

```python
def split_dataset(X, y, train_ratio=0.7, val_ratio=0.1):
    X_train, X_val, X_test = [], [], []
    y_train, y_val, y_test = [], [], []
    for a, b in zip(X, y):
        R = random.random()
        if 0 <= R <= train_ratio:
            X_train.append(a)
            y_train.append(b)
        elif train_ratio < R <= train_ratio + val_ratio:
            X_val.append(a)
            y_val.append(b)
        else:
            X_test.append(a)
            y_test.append(b)
    return (
        np.array(X_train),
        np.array(X_val),
        np.array(X_test),
        np.array(y_train),
        np.array(y_val),
        np.array(y_test),
    )
```

# Optimized KNN function using vectorized operations

```python
def knn_regression(X_train, y_train, X_val, y_val, k_values):
    errors = []
    for k in k_values:
        distances = np.sqrt(
            np.sum((X_val[:, None, :] - X_train[None, :, :]) ** 2, axis=2)
        )
        k_nearest_indices = np.argsort(distances, axis=1)[:, :k]

        k_nearest_outputs = y_train[k_nearest_indices]
        determined_outputs = np.mean(k_nearest_outputs, axis=1)

        error = np.mean((y_val - determined_outputs) ** 2)
        errors.append((k, error))
    return errors
```

```python
X = df_regression.drop(columns=["Rainfall"]).values
y = df_regression["Rainfall"].values
```

# Split the dataset

```python
X_train, X_val, X_test, y_train, y_val, y_test = split_dataset(X, y)
```

# K values to test

```python
k_values = [1, 3, 5, 10, 15]
```

# Calculate MSE for validation set

```python
validation_errors = knn_regression(X_train, y_train, X_val, y_val, k_values)
```

# Find the best K

```python
best_k, min_mse = min(validation_errors, key=lambda x: x[1])
print(f"Best K: {best_k}, Validation MSE: {min_mse}")
```

# Calculate MSE for test set using the best K

```python
test_errors = knn_regression(X_train, y_train, X_test, y_test, [best_k])
test_mse = test_errors[0][1]
print(f"Test MSE using best K ({best_k}): {test_mse}")
```

# Display the results in tabular form

```python
print("\nValidation Errors:")
print("K\tMean Squared Error")
for k, mse in validation_errors:
    print(f"{k}\t{mse}")
```

# Visualization of validation errors

```python
ks = [k for k, _ in validation_errors]
mses = [mse for _, mse in validation_errors]

plt.figure(figsize=(8, 6))
plt.plot(ks, mses, marker='o', label="Validation MSE")
plt.axvline(x=best_k, color='r', linestyle='--', label=f"Best K: {best_k}")
plt.title("KNN Regression: MSE vs. K")
plt.xlabel("Number of Neighbors (K)")
plt.ylabel("Mean Squared Error (MSE)")
plt.legend()
plt.grid(True)
plt.show()
```
